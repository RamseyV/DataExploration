Question 3.9.1
-------------------


```python
import pandas as pd

ages = [51,39,34,27,23,43,41,55,24,25,38,17,21,37,35,38,31,24,35,33]
ids = range(1,21)

df = pd.DataFrame()
df["ID"] = ids
df["AGE"] = ages
df
```

**a. Minimum, maximum and range:**

```python
minimum = df["AGE"].min()
maximum = df["AGE"].max()
age_range = maximum - minimum

print("Min:", minimum, "\nMax:", maximum, "\nRange:", age_range)
```

**b. Mean and median**

```python
mean = df["AGE"].mean()
median = df["AGE"].median()

print("Mean:", mean, "\nMedian:", median)
```

**c. Variance and standard deviation**

```python
std = df["AGE"].std()
variance = std**2

print("Variance:", variance, "\nStandard Deviation:", std)
```

**d. 1st quartile(25th percentile) and 3rd quartile (75th percentile)**

```python
q1 = df["AGE"].quantile(.25)
q3 = df["AGE"].quantile(.75)

print("1st Quartile:", q1, "\n3rd Quartile:", q3)
```

**e. Inter-quartile range**

```python
IQR = q3-q1
print("Inter-quartile range: ", IQR)
```

**f. 12th percentile**

```python
twelth = df["AGE"].quantile(.12)
print("12th percentile:", twelth)
```

Exploring Hate Crimes Dataset
===============

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/fivethirtyeight/data/master/hate-crimes/hate_crimes.csv'
df = pd.read_csv(url)
df
```

```python
df.info()
```

Let's remove the rows that don't have statistics about hate crimes. We may want the other feilds later on so we're going to copy the dataframe

```python
hate_crimes = df.dropna()

hate_crimes.info()
hate_crimes
```

```python
hate_crimes.describe()
```

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

scat_mat = scatter_matrix(hate_crimes, alpha=0.5, figsize=(15,15))

plt.show()
```

```python
hate_crimes.corr()
```

```python
plt.matshow(hate_crimes.corr())
```

Now let's take a look at the difference between hate_crimes_per_100k_splc (taken in 2016) and avg_hatecrimes_per_100k_fbi (taken from 2010-2015). *Hopefully this will be negative if the hate crimes go down.*

```python
diff = pd.DataFrame()
diff["state"] = hate_crimes["state"]
diff["difference"] = hate_crimes["hate_crimes_per_100k_splc"] - hate_crimes["avg_hatecrimes_per_100k_fbi"] 
diff
```

For these states it seems as though the hate crime rate decreases in all of them. 

Let's order the list and see which states had the highest and lowest decreases

```python
diff.sort_values("difference",ascending=False)
```

At this point it could be useful to see which states had the highest rates from 2010-2015...

```python
hate_crimes2010 = pd.DataFrame()
hate_crimes2010["state"] = hate_crimes["state"]
hate_crimes2010["avg_hatecrimes_per_100k_fbi"] = hate_crimes["avg_hatecrimes_per_100k_fbi"]
hate_crimes2010.sort_values("avg_hatecrimes_per_100k_fbi", ascending=False)
```
